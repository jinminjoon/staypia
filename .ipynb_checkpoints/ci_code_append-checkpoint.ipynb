```python
cicode_file_path=Desktop+"/staypia_city_데이터수집/mapped_city_list.xlsx"
cicode_data = pd.read_excel(cicode_file_path)
cicode_data
```

```python
cicode=list(cicode_data['ci_code'])
cdidx=list(cicode_data['cd_idx'])
```

```python
cicode_cdidx_dict=dict(zip(cdidx,cicode))
cicode_cdidx_dict
```

```python
len(cicode_cdidx_dict)
```

```python
for idx, data in tqdm(result_data[0:].iterrows(), desc="돌리는중"):
    result_data['ci_code'][idx]=cicode_cdidx_dict[data['cd_idx']]

print('finish')
```

```python
result_data
```

```python
columns = [
    'ci_code','cd_idx', 'cd_country', 'cd_name', 'link', 'title',
    'text', 'image', 'name',  'duration', 'regdate_ment', 'viewCount'
]
```

```python
result_df = pd.DataFrame(result_data, columns=columns)
result_df
```

```python
check_columns  = ['ci_code','cd_idx', 'cd_country', 'cd_name', 'link', 'title','image', 'name']
missing_values = result_df[result_df[check_columns].isnull().any(axis=1)]

print(len(result_df))
print(len(missing_values))
print(len(result_df) - len(missing_values))
```

```python
save_path=Desktop+f'/staypia_city_데이터수집/mapped_city_list_crawling_reseult_with_cicode.xlsx'
```
