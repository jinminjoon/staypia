```python
for idx, row in tqdm(db_youtube[0:].iterrows(), total = db_youtube[0:].shape[0], desc="돌리는중"):
    if '조회수 없음' in row['viewCount_raw']:
        db_youtube['viewCount'][idx]=0
    
    else:
        raw = row['viewCount_raw'].replace('조회수 ','').replace('회','')
        if '만' in raw:
            raw=float(raw.replace('만',''))
            raw=raw*10000
            db_youtube['viewCount'][idx]=int(raw)

        elif '천' in raw:
            raw=float(raw.replace('천',''))
            raw=raw*1000
            db_youtube['viewCount'][idx]=int(raw)

        else:
            db_youtube['viewCount'][idx]=int(raw)
        
```
