```python
old_temp_df = pd.DataFrame(data, columns=columns)
old_temp_df
```

# 교집합 찾기 - 실패 나중에 해보기

```python
del_merge=pd.merge(left = new_temp_df , right = old_temp_df, how = "inner", on = "cd_idx")
del_merge
```

# 반복문으로 중복제거해 새로운 도시 cd_idx 정보를 가진 리스트 생성

```python
old=list(old_temp_df['cd_idx'])
old
```

```python
new=list(new_temp_df['cd_idx'])
new
```

```python
serch_list=[]
```

```python
for idx in new: 
    if idx not in old: 
        serch_list.append(idx)
```

```python
len(serch_list)
```

# 새로운 검색 정보 파일 생성

```python
#수집 데이터 파일에 맞게 수정
result_columns = ['cd_idx','country_name','cd_name']
```

```python
data_list = [] 
```

```python
for idx, row in tqdm(data_csv[0:].iterrows(), total = data_csv[0:].shape[0], desc="돌리는중"):
    if row['cd_idx'] in serch_list:
        dataii = {"cd_idx": row['cd_idx'], "country_name": row["country_name"], "cd_name": row["cd_name"]}
        data_list.append(dataii)
```

```python
result_df = pd.DataFrame(data_list, columns=result_columns)
result_df
```
